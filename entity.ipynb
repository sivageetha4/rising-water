```python
| Field    | Data Type    | Key         |
| -------- | ------------ | ----------- |
| UserID   | INT          | Primary Key |
| Name     | VARCHAR(100) |             |
| Email    | VARCHAR(100) | Unique      |
| Password | VARCHAR(255) |             |
| Role     | VARCHAR(20)  |             |
| Field             | Data Type   |
| ----------------- | ----------- |
| ApplicantID       | INT (PK)    |
| UserID            | INT (FK)    |
| Gender            | VARCHAR(20) |
| Married           | VARCHAR(20) |
| Dependents        | VARCHAR(10) |
| Education         | VARCHAR(50) |
| SelfEmployed      | VARCHAR(20) |
| ApplicantIncome   | FLOAT       |
| CoapplicantIncome | FLOAT       |
| LoanAmount        | FLOAT       |
| LoanAmountTerm    | FLOAT       |
| CreditHistory     | INT         |
| PropertyArea      | VARCHAR(30) |
| Field     | Data Type    |
| --------- | ------------ |
| ModelID   | INT (PK)     |
| ModelName | VARCHAR(100) |
| Algorithm | VARCHAR(100) |
| Accuracy  | FLOAT        |
| ModelFile | VARCHAR(100) |
| Field          | Data Type   |
| -------------- | ----------- |
| PredictionID   | INT (PK)    |
| ApplicantID    | INT (FK)    |
| ModelID        | INT (FK)    |
| LoanStatus     | VARCHAR(20) |
| Probability    | FLOAT       |
| PredictionDate | DATETIME    |
Users
------
UserID (PK)
Name
Email
Password
Role
        |
        | 1
        |
        | N
Loan_Applicants
--------------------------
ApplicantID (PK)
UserID (FK)
Gender
Married
Dependents
Education
SelfEmployed
ApplicantIncome
CoapplicantIncome
LoanAmount
LoanAmountTerm
CreditHistory
PropertyArea
        |
        |1
        |
        |1
Prediction_Result
------------------------
PredictionID (PK)
ApplicantID (FK)
ModelID (FK)
LoanStatus
Probability
PredictionDate
        ^
        |
        |N
        |
ML_Model
-------------------
ModelID (PK)
ModelName
Algorithm
Accuracy
ModelFile
CREATE DATABASE loan_prediction;

USE loan_prediction;
CREATE TABLE Users(
    UserID INT AUTO_INCREMENT PRIMARY KEY,
    Name VARCHAR(100),
    Email VARCHAR(100) UNIQUE,
    Password VARCHAR(255),
    Role VARCHAR(20)
);
CREATE TABLE Loan_Applicants(
    ApplicantID INT AUTO_INCREMENT PRIMARY KEY,
    UserID INT,
    Gender VARCHAR(20),
    Married VARCHAR(20),
    Dependents VARCHAR(10),
    Education VARCHAR(50),
    SelfEmployed VARCHAR(20),
    ApplicantIncome FLOAT,
    CoapplicantIncome FLOAT,
    LoanAmount FLOAT,
    LoanAmountTerm FLOAT,
    CreditHistory INT,
    PropertyArea VARCHAR(30),

    FOREIGN KEY(UserID)
    REFERENCES Users(UserID)
);
CREATE TABLE ML_Model(
    ModelID INT AUTO_INCREMENT PRIMARY KEY,
    ModelName VARCHAR(100),
    Algorithm VARCHAR(100),
    Accuracy FLOAT,
    ModelFile VARCHAR(100)
);
CREATE TABLE Prediction_Result(
    PredictionID INT AUTO_INCREMENT PRIMARY KEY,

    ApplicantID INT,

    ModelID INT,

    LoanStatus VARCHAR(20),

    Probability FLOAT,

    PredictionDate DATETIME,

    FOREIGN KEY(ApplicantID)
    REFERENCES Loan_Applicants(ApplicantID),

    FOREIGN KEY(ModelID)
    REFERENCES ML_Model(ModelID)
);
INSERT INTO ML_Model
(ModelName, Algorithm, Accuracy, ModelFile)

VALUES

('Loan Approval Model',
 'Random Forest',
 86.5,
 'rdf.pkl');
 INSERT INTO Users
(Name,Email,Password,Role)

VALUES

('Admin',
'admin@gmail.com',
'admin123',
'Admin');
Smart-Lender/
│
├── app.py
├── rdf.pkl
├── loan_prediction.csv
├── requirements.txt
│
├── templates/
│      home.html
│      predict.html
│      submit.html
│
├── static/
│      style.css
│
└── database/
       loan_prediction.sql
       Dataset
      ↓
Data Preprocessing
      ↓
Train ML Models
      ↓
Random Forest Selected
      ↓
Save Model (rdf.pkl)
      ↓
Flask Application
      ↓
User Inputs Loan Details
      ↓
Load rdf.pkl
      ↓
Prediction
      ↓
Display Loan Approved / Rejected
      ↓
Store Prediction in Database
```
